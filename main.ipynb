### • Plotting a Histogram of Age :

```python
age_min = df['Age'].min()
age_max = df['Age'].max()
bin_width = 2
num_bins_age = (age_max - age_min) // bin_width

df['Age'].hist(bins=num_bins_age , edgecolor='black')

plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
# plt.grid(False)
plt.show()
```

```python
mean_age = df['Age'].mean()      
median_age = df['Age'].median()  
mode_age = df['Age'].mode()[0]    
age_range = df['Age'].max() - df['Age'].min()
variance_age = df['Age'].var()
std_dev_age = df['Age'].std() 

print(f"Mean Age: {mean_age.__round__(0)}")
print(f"Median Age: {median_age.__round__(0)}")
print(f"Mode Age: {mode_age}")
print(f"Range of Age: {age_range}")
print(f"Variance of Age: {variance_age.__round__(0)}")
print(f"Standard Deviation of Age: {std_dev_age.__round__(0)}")
```

##### First, we will perform an analysis between the categorical variables 'Result' and 'Gender' :

```python
grouped_data = df.groupby(['Gender', 'Result']).size().unstack()
print(grouped_data)
```

##### Analysis between the Numbercal variables 'Heart Rate' and 'Blood Sugar' :

Scatter Plot between Heart Rate and Blood Sugar

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Blood sugar', y='Heart rate', data=df)

plt.xlabel('Blood Sugar')
plt.ylabel('Heart Rate')
plt.title('Scatter Plot of Blood Sugar vs Heart Rate')

plt.show()
```

Correlation and Covariance :


```python
correlation = df['Blood sugar'].corr(df['Heart rate'])
print(f"Correlation between Blood Sugar and Heart Rate: {correlation:.2f}")

covariance = df['Blood sugar'].cov(df['Heart rate'])
print(f"Covariance between Blood Sugar and Heart Rate: {covariance:.2f}")
```

Interpretation:

(i) From the analysis of the data plot, it's observed that there is a very weak positive association between the variables being considered, which are Heart Rate and Blood Sugar levels. 

(ii) The very weak correlation (0.01) suggests that Blood Sugar and Heart Rate are not strongly related in a linear sense. The positive covariance suggests that there is some tendency for Heart Rate to increase when Blood Sugar increases, but the lack of a strong correlation indicates this relationship is weak and not necessarily significant.

##### Using Hinton Diagram to find any relation between the variables: 

```python
def hinton(matrix, max_weight=None, ax=None):
    ax = ax if ax is not None else plt.gca()
    if not max_weight:
        max_weight = 2**np.ceil(np.log(np.abs(matrix).max())/np.log(2))
    ax.patch.set_facecolor('lightgray')
    ax.set_aspect('equal', 'box')
    ax.xaxis.set_major_locator(plt.NullLocator())
    ax.yaxis.set_major_locator(plt.NullLocator())
    for (x, y), w in np.ndenumerate(matrix):
        color = 'red' if w > 0 else 'blue'
        size = np.sqrt(np.abs(w))
        rect = plt.Rectangle([x - size / 2, y - size / 2], size, size,
        facecolor=color, edgecolor=color)
        ax.add_patch(rect)
    nticks = matrix.shape[0]
    ax.xaxis.tick_top()
    ax.set_xticks(range(nticks))
    ax.set_xticklabels(list(matrix.columns), rotation=90)
    ax.set_yticks(range(nticks))
    ax.set_yticklabels(matrix.columns)
    ax.grid(False)
    ax.autoscale_view()
    ax.invert_yaxis()
hinton(numerical_attributes.corr())
```

Interpretation:

(i) We can observe a correlation between systolic blood pressure (SBP) and diastolic blood pressure (DBP), which is to be expected and does not provide any new insights.

(ii) There is a small positive correlation between heart rate and diastolic blood pressure (DBP), and a negative correlation between heart rate and systolic blood pressure (SBP). This suggests that heart rate affects blood pressure, which is a valid conclusion.
